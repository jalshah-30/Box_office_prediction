```python
import pandas as pd
```

```python
data=pd.read_csv('Movies.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(['imdb_id','title','release_date','overview','release_year'],axis=1,inplace=True)
```

```python
data.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#missing Values
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data["collection_crore"], bins=40, kde=True)
plt.title("Collection Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data["budget_crore"], bins=40, kde=True)
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x="budget_crore",
    y="collection_crore",
    data=data
)

plt.show()
```

```python
numeric = data.select_dtypes(include=np.number)

plt.figure(figsize=(8,6))

sns.heatmap(
    numeric.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.show()
```

```python
top = data["directors"].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top.values,
    y=top.index
)

plt.title("Top Directors")
plt.show()
```

```python
top = data["cast"].value_counts().head(15)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top.values,
    y=top.index
)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    y=data["genres"],
    order=data["genres"].value_counts().index
)

plt.show()
```
