```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("indian_movies.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df=df[:1500]
```

```python
df.info()
```

```python
df=df.dropna()
df=df.drop_duplicates()
```

```python
df.info()
```

```python
df.drop(['imdb_id','title','vote_average','release_date','overview','production_companies','release_year'],axis=1,inplace=True)
```

```python
df["budget"] = df["budget"].astype(float).astype(int)
df["collection"]=df["collection"].astype(float).astype(int)
```

```python
df.info()
```

```python
##Unique values
for col in df.columns:
    print(col)
    print(df[col].nunique())
    print()
```

```python
num_cols = ["vote_count","popularity","budget","collection"]

for col in num_cols:
    plt.figure(figsize=(7,4))
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(6,2))
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

```python
df["original_language"].value_counts()
```

```python
sns.countplot(data=df,x="original_language")
plt.show()
```

```python
genre_counts = (
    df["genres"]
    .str.split(", ")
    .explode()
    .value_counts()
)

genre_counts
```
